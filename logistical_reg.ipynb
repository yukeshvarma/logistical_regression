```python
df=df.drop(columns=['passenger_id','name','ticket','cabin','home.dest','boat'])
```

```python
df.head()
```

```python
df=df.fillna(df['age'].mode())
```

```python
df=df.fillna(df['embarked'].mode())

df=df.fillna(df['body'].median())
```

```python
df.isnull().sum()
```

```python
df=pd.get_dummies(df,columns=['sex','embarked'])
```

```python
df.info()
```

```python
dfx=df.drop(columns=['survived'])
```

```python
x=dfx.iloc[:,:]
y=df.iloc[:,6:7].values
```

```python
import numpy as np
```

```python
x.head()
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=111,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
log =LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
y_predicted = log.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,recall_score,accuracy_score,f1_score,precision_score
cm = confusion_matrix(y_test,y_predicted)
print('confusion matrix',cm)
re=recall_score(y_test,y_predicted)
print('recall score',re)
ac=accuracy_score(y_test,y_predicted)
print('accuracy score',ac)
f1=f1_score(y_test,y_predicted)
print('f1 score',f1)
pre =precision_score(y_test,y_predicted)
print('precision score',pre)
```
